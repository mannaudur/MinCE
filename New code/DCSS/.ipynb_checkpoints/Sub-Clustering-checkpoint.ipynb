## Deterministic Column Subset Selection

```python
# Lifted from https://github.com/srmcc/dcss_single_cell/blob/master/src/dcss/dls_funct.py

import pandas as pd
import numpy as np

def det_leverage(V, k, epsilon):
    """
    for the data matrix A=U \Sigma V^T
    A.shape =(n=number of samples, d=number of features)
    V.shape=(d, n)
    k is the rank of the PCA leverage score
    epsilon is the error parameter.
    the function returns
    theta: the number of kept columns
    index_keep: the index of the selected columns
    tau_sorted:  the sorted leverage scores of all the columns
    index_drop:  the index of the dropped columns
    """
    V = pd.DataFrame(V)
    Vk= V.iloc[:, 0:k]
    print(Vk.shape)
    tau= np.sum(Vk**2, axis=1)
    tau_sorted= tau.sort_values(ascending=False, inplace=False)
    lev_sum=0
    for i in range(V.shape[0]):
        lev_sum =lev_sum+ tau_sorted.iloc[i]
        if lev_sum > k - epsilon:
            theta=i+1
            if theta >= k:
                break
    index_keep = tau_sorted.index[0:theta]
    index_keep = index_keep.values
    index_drop = tau_sorted.index[theta:]
    index_drop = index_drop.values
    return(theta, index_keep, tau_sorted, index_drop)

def read_bitmatrix(path):
    bitmat = pd.read_csv(path, sep='\t').values[:,1:]
    #valid = np.where(np.sum(self.bitmat, axis=1)<int(np.shape(bitmat)[1]))[0]
    return(bitmat)
```

```python
import numpy as np

def load_sketch(path):
    with open(path) as f:
        lines = f.readlines()
        name = lines[0].split('/')[-1]
        sketch = set(lines[4::])
        return(name, sketch)
    
def load_paths(ref_path):
    with open(ref_path) as f:
        lines = f.readlines()
        return(lines)

def distMat(clique_path):
    paths = load_paths(clique_path)
    n = len(paths)
    D = np.zeros((n,n))
    names = []
    for i in range(n):
        name, sketch = load_sketch(paths[i].strip())
        names.append(name.strip())
        for j in range(n):
            D[i,j] = 5000 - len(sketch.intersection(load_sketch(paths[j].strip())[1]))
    return(D,names)
```

```python
def homeMade(D, thresh):
    valid = np.ones(D.shape[0])
    max_ = np.max(D)+1
    king, queen = np.unravel_index(D.argmax(), D.shape)
    D_king = np.matrix.copy(D[king,:])
    D_queen = np.matrix.copy(D[queen,:])
    clique1 = set()
    clique2 = set()
    while(np.sum(valid)):
        closest_to_center = D_king.argmin()
        clique1.add(closest_to_center)
        D_king[closest_to_center] = max_
        D_queen[closest_to_center] = max_
        valid[closest_to_center] = 0
        closest_to_center = D_queen.argmin()
        clique2.add(closest_to_center)
        D_queen[closest_to_center] = max_
        D_king[closest_to_center] = max_
        valid[closest_to_center] = 0
    c1_fixed = clique1.copy()
    c2_fixed = clique2.copy()
    for mem in clique1:
        close = np.where(D[mem,:] <= thresh)
        c1_fixed.update(close[0])
    for mem in clique2:
        close = np.where(D[mem,:] <= thresh)
        c2_fixed.update(close[0])
            
    return(c1_fixed, c2_fixed, clique2, clique1)
        
    
def simple_cliques(D, thresh):
    mems = np.arange(D.shape[0])
    cliques = set()
    singles = set()
    for mem in mems:
        merge = np.where(D[mem,:] <= thresh)[0]
        if len(merge)>1:
            cliques.add(tuple(merge))
        else:
            singles.add(merge[0])
            
    print(len(singles), " singletons found.")
    print(len(cliques), " cliques found:")
    for clique in cliques:
        print("Size:", len(clique), "-", clique)
    return(cliques, mems, singles)
                
    
def simple_cliques_with_subset_merge(D, ids, thresh):
    mems = np.arange(D.shape[0])
    cliques = list()
    singles = set()
    for mem in mems:
        merge = ids[np.where(D[mem,:] <= thresh)[0]]
        if len(merge)>1:
            merge = set(merge)
            add_ = True
            for clq in cliques:
                if clq.issubset(merge) or merge.issubset(clq):
                    clq.update(merge)
                    add_ = False
            if add_:
                cliques.append(merge)
        else:
            singles.add(merge[0])
                    
    print(len(singles), " singletons found.")
    print(len(cliques), " cliques found:")
    for clique in cliques:
        print("Size:", len(clique), "-", clique)
    return(cliques, singles)
```

```python
def count_0_dist(D):
    zeros = (len(np.where(D == 0)[0]) - D.shape[0])/2
    total = ((D.shape[0]**2) - D.shape[0])/2
    
    return zeros, total
```

### Graph visualization and Max flow / min cut

```python
import networkx as nx
import matplotlib.pyplot as plt

def makeGraph(D, draw):
    G = nx.DiGraph()

    for i in range(D.shape[0]):
        for j in range(D.shape[0]):
            if i != j:
                G.add_edge(i, j, capacity= D[i,j], weight = -D[i,j])
    if draw:
        nx.draw_networkx(G)
    return(G)
```

```python
def MC_partition(D, thresh, draw):
    G = makeGraph(D, False) #ATH THRESHOLD FOR CLIQUE JOINING SET TO 5 !!!
    if(D.shape[0]<2):
        return(False, set([tuple(G.nodes)]), set([tuple(G.nodes)])) #Kind of ugly, just making this work...
    success = False
    sink, source = np.unravel_index(D.argmax(), D.shape)
    cut_value, partition = nx.minimum_cut(G, sink, source)
    reachable, non_reachable = partition
    print(reachable)
    print(non_reachable)
    cutset = set()
    for u, nbrs in ((n, G[n]) for n in reachable):
        cutset.update((u, v) for v in nbrs if v in non_reachable)
    for edge in sorted(cutset):
        if D[edge[0], edge[1]] <= 5:
            reachable.add(edge[0])
            non_reachable.add(edge[0])
            reachable.add(edge[1])
            non_reachable.add(edge[1])
    print(cutset)
    
    # Checking that new partition is sufficiently large on both ends   
    thresh_check = (abs(len(reachable) - len(non_reachable)) <= D.shape[0]*thresh)
    # Checking that we actually partitioned something, once we split nodes to both parent cliques
    iter_check =  (len(non_reachable) < D.shape[0] and len(reachable) < D.shape[0])
    if thresh_check and iter_check:
        success = True
    
    print("Threshold:", thresh_check)
    print("Iter: ", iter_check)
    if draw:
        nx.draw_networkx(G)
        plt.show()
    return(success, list(reachable), list(non_reachable))

def iter_MC(G, D, thresh):
    cliques = set() #empty set of sub-cliques
    G_ = MC_partition(D, thresh, False) #make partition into two parts
    if(G_[0]): #if partition is successful
        G_1 = iter_MC(G_, D[np.ix_(G_[1],G_[1])], thresh) #go into first sub-clique for partition attempt
        G_2 = iter_MC(G_, D[np.ix_(G_[2],G_[2])], thresh) #then second sub-clique for partition attempt
        cliques.update(G_1) #add both results to set of sub-cliques
        cliques.update(G_2)
        return(cliques) #return list of sub-cliques for further adding
    
    cliques.update(tuple(G[1])) #otherwise add original results to set of sub-cliques
    cliques.update(tuple(G[2]))
    return(cliques) #return list of sub-cliques for further adding
    
# D is a distance matrix for the entire clique
# b <= 0.5 is the threshold of the smallest split that qualifies for further partitioning. 
# Example: b=0.2 would mean that a partition of the clique with less than 20% in the
# smaller cluster would end iteration and return the full clique.
def get_subcliques(D, b):
    clique_set = set()
    G = MC_partition(D, 1 - 2*b, False)
    if(G[0]):
        clique_set = iter_MC(G, D, 1 - 2*b)
        print("Iterated...")
    else:
        clique_set = G[1]
    return(clique_set)

# Here I might need to process the clique_set to fold subsets of cliques together
# 
```

### Multi-dimensional scaling

```python
from sklearn.manifold import MDS

def MDS_transform(D, n_init_val, max_iter_val):
    min_stress = float('inf')
    best_n = 0
    best_embedding = D
    for n in range(2,D.shape[0]+1):
        mds = MDS(n_components=n, dissimilarity='precomputed', random_state=0, 
                  n_init = n_init_val, max_iter = max_iter_val)
        # Get the embeddings
        X_transform_L1 = mds.fit(D)
        if mds.stress_ < min_stress:
            min_stress = mds.stress_
            best_n = n
            best_embedding = mds.embedding_

    return(best_embedding, best_n, min_stress)
```

## Bottom Up sub-clustering

```python
# Creates a distance matrix from a path to .clique file. This .clique file has
# relative paths to every sketch that was categorized into the clique by the 
# initial union-find method.

import numpy as np

def load_sketch(path):
    with open(path) as f:
        lines = f.readlines()
        name = lines[0].split('/')[-1]
        sketch = set(lines[4::])
        return(name, sketch)
    
def load_paths(ref_path):
    with open(ref_path) as f:
        lines = f.readlines()
        return(lines)
    
def read_indices_to_rev_dict(path):
    indices = load_paths(path)
    rev_indices = dict()
    for line in indices:
        entries = line.split(" ")
        rev_indices[entries[1]] = entries[0]
    return(rev_indices)
    
def distMat(clique_path):
    paths = load_paths(clique_path)
    n = len(paths)
    D = np.zeros((n,n))
    names = []
    for i in range(n):
        name, sketch = load_sketch(paths[i].strip())
        names.append(name.strip())
        for j in range(n):
            D[i,j] = 5000 - len(sketch.intersection(load_sketch(paths[j].strip())[1]))
    return(D,names)

# Takes in a distance matrix D_, which starts out as the original distance matrix
# but with each iteration we work with a submatrix of D, corresponding to the
# sketches yet to be assigned to a sub-clique. The parameter ids is a mapping between
# the indices of the orignal distance matrix D and the new submatrix D_.
# Thresh refers to the threshold distance value to unite sketches into cliques
# and the reference matrix is a table showing where each sketch ended up, which
# is persistently appended throughout the iterative process (that is between runs
# of this algorithm)
def simple_cliques_with_subset_merge(D, ids, thresh, reference_matrix):
    mems = np.arange(D.shape[0]) #members in the submatrix
    cliques = list()
    singles = set()
    for mem in mems:
        merge = ids[np.where(D[mem,:] <= thresh)[0]] #Using ids to get original indices for accounting
        if len(merge)>1: #checking if there is any match within the designated threshold
            merge = set(merge)
            add_ = True
            for clq in cliques:
                if clq.issubset(merge) or merge.issubset(clq): #Check if we have a superset in the log
                    clq.update(merge) #merge subset with superset
                    add_ = False #don't make a new one
            if add_:
                cliques.append(merge) #add a new set
        else:
            singles.add(merge[0]) #adding to singletons if no match
    print("For threshold", thresh, ":")
    print(len(singles), " singletons found.")
    print(len(cliques), " cliques found")
    for i, clique in enumerate(cliques):
        print("Size:", len(clique), "-", clique)
        for mem in clique:
            #log the position of each member for reference
            #thresh refers to line# in ref_matrix and i refers to # in list of cliques for that threshold
            #for example, if ref_mat[k,j] = m, then sketch j was joined to clique m with threshold k.
            reference_matrix[thresh, mem] = i 
    return(cliques, singles, reference_matrix)

# Organizing step in between runs of simple_cliques_with_subset_merge()
def clean_up_cliques(D, ids_):
    ids = np.array(list(ids_)) #convert singletons to new array of ids
    D_ = D[np.ix_(ids,ids)] #create submatrix of original distance matrix D, pertaining to those singleton ids
    return(D_, ids)

# Runs iterative calls on simple_cliques_with_subset_merge() with ever larger threshold values
# up to and including iter_. The path is just the path to give to distMat. The clean_up paramter
# is a boolean designating whether to return singletons that couldn't be placed in a clique in 
# the given amount of iter_ (clean_up=False) or to iterate over left over singletons and the end
# of iteration and assign them to the cliques containing their closest pairing in the distance matrix
def bottom_up_cliques(path, iter_, clean_up = True):
    D, names = distMat(path) #Make
    reference_matrix = np.ones((iter_+2,D.shape[0]))*-1 #Reference matrix with extra line for unassigned singletons
    cliques = []
    #clq is threefold: clq[0] is cliques found, clq[1] is singletons and clq[2] is updated reference matrix
    clq = simple_cliques_with_subset_merge(D,np.arange(D.shape[0]),0, reference_matrix)
    # cliques.append(clq[0:2]) #If we wanted singletons as well
    cliques.append(clq[0]) #Stowing cliques away for reference
    reference_matrix = clq[2] #Filling in the reference matrix
    for i in range(iter_):
        D_,ids = clean_up_cliques(D, clq[1]) #Clean up stage between runs
        clq = simple_cliques_with_subset_merge(D_,ids,i+1, reference_matrix)
        # cliques.append(clq[0:2]) #If we wanted singletons as well
        cliques.append(clq[0])
        reference_matrix = clq[2]
        if len(clq[1]) == 0: #if no returned singletons, the algorithm has finished
            print("--- Reached completion at threshold iteration",i,"---")
            clean_up = False
            break
    singletons = ids #Just a renaming for clarifying purposes
    if(clean_up):
        for singleton in singletons: #for each singleton left
            D[singleton,singleton] = float('inf') #we wanna find the actual closest match
            best_fit = np.argmin(D[singleton,:]) #which tells us which member to look for
            # there's a change the best_fit is a singleton, though it's very unlikely
            clique_thresh = np.argmax(reference_matrix[:,best_fit]) #which was put into a clique with this threshold
            clique_id = reference_matrix[clique_thresh,best_fit].astype(np.int64) #and this is its location in the list
            cliques[clique_thresh][clique_id].add(singleton) #so we add that singleton to its closest neighbour's clique
            reference_matrix[clique_thresh,singleton] = clique_id #and log it so placed
            print("Added singleton", singleton,"to clique no.",clique_id,"with threshold",clique_thresh)
            print("Amended form of clique is", cliques[clique_thresh][clique_id])
            
    else:
        for singleton in singletons:
            reference_matrix[iter_+1,singleton] = -2 #if clean_up = False, just log those singletons as unassigned
                
            
    return(cliques, names)

def write_cliques_to_disk(cliques_, cluster_path):
    cluster_id = cluster_path.split('/')[-1][:-7]
    for i in range(len(cliques_)):
        if(cliques_[i]):
            for j in range(len(cliques_[i])):
                members = cliques_[i][j]
                clique_id = cluster_id+"_"+i+"_"+j+".clique"
                with open('cliques/'+clique_id, 'w') as f:
                    for mem in members:
                        f.write(mem)
                        
def clique_every_cluster(all_file):
    for cluster_path in all_file:
        cliques_ = bottom_up_cliques(cluster_path, 4)
        
                        
```
